# Ejercicios I-O

**Example**

**Trade matrix (Z)**

   - $Z$ is a square matrix representing the flow  between sectors.
   - Each element $z_{ij}$ represents the value of goods or services produced by sector $i$ and consumed by sector $j$.


   $
   Z = 
   \begin{bmatrix}
   z_{11} & z_{12} & ... & z_{1n} \\
   z_{21} & z_{22} & ... & z_{2n} \\
   ... & ... & ... & ... \\
   z_{n1} & z_{n2} & ... & z_{nn}
   \end{bmatrix}
   $
   
   
**Value added vector (V)**

$
V = 
\begin{bmatrix}
va_{1} & va_{2} & ... & va_{n}
\end{bmatrix}
$



**Final demand vector (Y) of products purchased by final consumers**

$Y = 
\begin{bmatrix}
fd_{1} \\
fd_{2} \\
... \\
fd_{n} 
\end{bmatrix}
$

   
**Total output vector (X)**


Calculate total outputs: intermediate output + final demand (sum rows)

$X_i = \sum_{i=0}^n Z_i + Y_i$
    
Calculate the total inputs: intermediate input + value added (sum cols)

$X_j = \sum_{j=0}^n Z_j + V_j$

To calculate the **technical coefficient matrix**  $ A  = Z \cdot \hat{X}^{-1}$

Where:
- $ Z $: Inter-industry flow matrix.
- $\hat{X}^{-1}$: Diagonalized inverse of the output vector $ X $.


   - $\hat{X}$ is a diagonal matrix where the diagonal elements are the total outputs of each sector $ X_i $, and all off-diagonal elements are zero:
   
     $
     \hat{X} = 
     \begin{bmatrix}
     x_1 & 0   & 0   \\
     0   & x_2 & 0   \\
     0   & 0   & x_3
     \end{bmatrix}
     $
   - The inverse $\hat{x}^{-1}$ is calculated as:
   
     $
     \hat{X}^{-1} = 
     \begin{bmatrix}
     \frac{1}{x_1} & 0          & 0          \\
     0             & \frac{1}{x_2} & 0      \\
     0             & 0          & \frac{1}{x_3}
     \end{bmatrix}
     $


To calculate **Leontief Inverse ($L$) (B)**:

Use the formula:
$$
L_p = (I - A)^{-1}
$$

The Leontief Inverse represents the total (direct and indirect) effects on production across sectors caused by a unit increase in final demand.

---

To calculate **Leontief Inverse in Price Model ($L_p$) (B)**:



$\mathbf{L} = (\mathbf{I}-\mathbf{A}')^{-1}$

The Leontief Price Model shows how prices in different sectors depend on the input costs.

---

To calculate ** Value-Added Coefficients ($V_{\text{coeff}}$)** divide the value-added vector ($VA$) by the total output vector ($X$):

$$
V_{\text{coeff}} = \frac{VA}{X}
$$

---

To calculate **Price Vector ($p$)** multiply the Leontief inverse in the price model ($L_p$) with the value-added coefficients ($V_{\text{coeff}}$) to calculate the price vector:

$$
p = L_p \cdot V_{\text{coeff}}
$$

---

**Normalization Check:**

Verify that the price vector is normalized ($p = 1$) or check for the expected normalized values.

````python
# Load sample data WWZ

objIO = it.obj_IO(it.load_Test("WWZ"))
X = objIO["X"]
Z = objIO["Z"]
VA = objIO["VA"]
Y = objIO["Y"]
B = objIO["B"]
W = objIO["W"]

CS = objIO["dims"]["CS"]

cs_names = objIO["names"]["cs_names"]
c_names  = objIO["names"]["c_names"]


objIO
````

````python
# Technical coefficientes and leontief

# Step 1
x_hat_inv = np.linalg.pinv(np.diag(X)) #pseudo inversa por tener ceros

# Step 2: Compute the technical coefficient matrix A
A = Z @ x_hat_inv  # Use @ for matrix multiplication
# Equivalente el Step 2 a:
# A = io["Z"].values/io["X"].replace(0, np.nan).T.values 
# A = A.astype(float)
# A[~np.isfinite(A)] = 0


# Calculate the leontief inverse matrix in the Demand-pull model
# Step 1: Calculate (I - A)
I = np.identity(CS)
I_minus_A = I - A

# Step 2: Calculate the Leontief inverse L
L = np.linalg.inv(I_minus_A)


# Leontief inverse price model
# Step 1: Calculate the Leontief inverse matrix in the price model
L_p = np.linalg.pinv(np.identity(CS) - A.T)  # (I - A')^-1

# Step 2: Calculate the value-added coefficients (v_coeff)
v_coeff = (VA.squeeze() / X).squeeze()  # Element-wise division of v by x_out

# Step 3: Calculate the price vector by multiplying the Leontief inverse with the value-added coefficients
p = L_p @ np.nan_to_num(v_coeff, nan=0)
p
````

````python
# ============================================================
# Technical coefficients and Leontief models (Quantity model and Price model)
# ============================================================

# ------------------------------------------------------------
# 1. Technical coefficients matrix (A)
# ------------------------------------------------------------
# A = Z · X̂⁻¹
# Coeficientes técnicos: inputs intermedios por unidad de output
# ------------------------------------------------------------

# Inverse of diagonal output matrix (robust to zeros)
x_hat_inv = np.linalg.pinv(np.diag(X))

# Technical coefficient matrix
A = Z @ x_hat_inv


# ------------------------------------------------------------
# 2. Leontief inverse – Quantity model (Demand-pull)
# ------------------------------------------------------------
# x = (I − A)⁻¹ · y
# ------------------------------------------------------------

# Identity matrix
I = np.eye(CS)

# Net requirements matrix
I_minus_A = I - A

# Leontief inverse (quantity model)
L = np.linalg.inv(I_minus_A)


# ------------------------------------------------------------
# 3. Leontief inverse – Price model (Cost-push)
# ------------------------------------------------------------
# p = (I − A')⁻¹ · v
# ------------------------------------------------------------

# Leontief inverse in the price model
L_p = np.linalg.pinv(np.eye(CS) - A.T)


# ------------------------------------------------------------
# 4. Value-added coefficients
# ------------------------------------------------------------
# v_i = VA_i / X_i
# Coste primario por unidad de producción
# ------------------------------------------------------------

v_coeff = np.divide(VA.squeeze(), X, out=np.zeros_like(X, dtype=float), where=X != 0)


# ------------------------------------------------------------
# 5. Equilibrium price vector
# ------------------------------------------------------------
# p = L_p · v
# ------------------------------------------------------------

p = L_p @ v_coeff


# ------------------------------------------------------------
# 6. (Optional) Consistency check – Price system
# ------------------------------------------------------------
# p = A' · p + v
# ------------------------------------------------------------

assert np.allclose(
    p,
    A.T @ p + v_coeff,
    atol=1e-6
), "Chequeo del sistema de precios falló"


# Mostrar resultados principales
print("=== Technical coefficients and Leontief models ===")
print("Shape of A:", A.shape)
print("Shape of L (quantity model):", L.shape)
print("Shape of L_p (price model):", L_p.shape)
print("First elements of price vector p:")
print(p[:10])
````

````python
# ============================================================
# Consistency checks: Output identity in the Leontief model
# ============================================================
# Objective:
# Verify that total output X is consistent with:
#   (i) the Leontief quantity model: X = L · Y
#   (ii) intermediate + final uses: X = Z·1 + Y·1
# ============================================================


# ------------------------------------------------------------
# 1. Check using the Leontief inverse (Quantity model)
# ------------------------------------------------------------
# X = L · y
# where y is total final demand aggregated by sector
# ------------------------------------------------------------

# Aggregate final demand
y = Y.sum(axis=1)

# Leontief-based output
print("Opt.1 – X = L·Y  (residuals):", np.round(L @ y - X, 5))


# ------------------------------------------------------------
# 2. Check using induced output from B · Y
# ------------------------------------------------------------
# B = (I − A)⁻¹  ≡ Leontief inverse
# Sum across destination countries
# ------------------------------------------------------------

BY = B @ Y

print("Opt.2 – X = sum(B·Y)  (residuals):", np.round(BY.sum(axis=1) - X, 5))

# (optional) explicit output vector
X_from_BY = BY.sum(axis=1)
pd.DataFrame(X_from_BY, columns=["X_from_BY"]).head()


# ------------------------------------------------------------
# 3. Check using uses-side accounting identity
# ------------------------------------------------------------
# X = intermediate uses + final demand
#   = Z·1 + Y·1
# ------------------------------------------------------------

print("Opt.3 – X = Z + Y  (residuals):", np.round(np.sum(np.hstack((Z, Y)), axis=1) - X, 5) )


# ------------------------------------------------------------
# 4. Check consistency between Z and A·X
# ------------------------------------------------------------
# By definition:
#   Z = A · X̂
# ------------------------------------------------------------

print("Opt.4 – Z = A·X  (total residual):",  np.round((Z - A * X).sum(), 6))
````

````python
# ============================================================
# 1. Output inducido por la demanda final
# ============================================================
# BY = B · Y
#
# B  : inversa de Leontief (I − A)^−1
# Y  : demanda final
#
# Interpretación:
# Producción total (directa + indirecta) necesaria para satisfacer
# una unidad de demanda final en cada país.
#
# Dimensión:
# BY → (sector–país) × (país demandante)
# ============================================================


BY = B @ Y

df_BY = pd.DataFrame(BY.sum(axis=1), index=cs_names, columns=["BY"])

df_X = pd.DataFrame(X, index=cs_names, columns=["X"])

# Totales agregados
total_BY = df_BY["BY"].sum()
total_X  = df_X["X"].sum()

print("=== Output inducido por demanda final ===")
print(f"Output total inducido (Σ BY): {total_BY:,.3f}")
print(f"Output total observado (Σ X): {total_X:,.3f}\n")

# Inspección rápida
print(pd.concat([df_BY.head(10), df_X.head(10)], axis=1))


# ============================================================
# 2. Valor añadido inducido por la demanda final
# ============================================================
# WBY = W · B · Y
#
# W : matriz diagonal de coeficientes de valor añadido (VA / X)
#
# Interpretación económica:
# Valor añadido generado en cada sector–país como consecuencia
# de la demanda final de cada país.
#
# Dimensión:
# WBY → (sector–país productor) × (país demandante)
# ============================================================

WBY = W @ B @ Y

df_WBY = pd.DataFrame( WBY, index=cs_names, columns=c_names)

VA_total = df_WBY.values.sum()

print("\n=== Valor añadido inducido por demanda final ===")
print(f"VA total inducido (Σ WBY): {VA_total:,.3f}")

display(df_WBY.head(5))


# ============================================================
# 3. Valor añadido absorbido en el extranjero
# ============================================================
# Objetivo:
# Eliminar el VA doméstico y quedarnos solo con el VA generado
# en un país y absorbido por la demanda final de OTROS países.
#
# Estrategia:
# - Extraer el código país del productor (filas)
# - Extraer el código país del demandante (columnas)
# - Construir una máscara país-productor == país-demandante
# - Poner a cero esos elementos (VA doméstico)
# ============================================================

row_country = pd.Index(cs_names).str[:3].to_numpy()
col_country = pd.Index(c_names).str[:3].to_numpy()

mask_domestic = row_country[:, None] == col_country[None, :]

df_WBY_abroad = df_WBY.mask(mask_domestic, 0)

VA_abroad = df_WBY_abroad.values.sum()

print("\n=== Valor añadido absorbido en el extranjero ===")
print(f"VA absorbido abroad (Σ WBY_abroad): {VA_abroad:,.3f}")
print(f"Share del VA abroad (%): {100 * VA_abroad / VA_total:,.2f}")

display(df_WBY_abroad.head(5))


# ============================================================
# 4. Test contable básico
# ============================================================

print("\n=== Chequeo contable ===")

assert np.allclose((W @ BY).sum(), VA_total, atol=1e-6), "Chequeo W·BY vs W·B·Y falló"

print("Chequeo W · BY == W · B · Y : OK ✔️")
````

### Exercise: 1 Country, 3 Sector Numerical IO Table

````python

# # https://gist.github.com/floorbroekman
# Dimensiones
C = 1
S = 3
FD = 1
CS = C*S
CFD = C*FD

# CREACIÓN DE NOMBRES
c_names = ["1"]
s_names = ["Agr", "Man", "Ser"]
cs_names = ["1_Agr", "1_Man", "1_Ser"]
fd_names = [""]
cfd_names = ["1_0"]


labels = pd.Index(["Agriculture", "Manufacturing", "Services"])

Z = pd.DataFrame([
    [0.6,2.6,0.5],
    [0.8, 30.6, 7.8],
    [0.9,12.1,23]], index=labels, columns=labels)


# After having looked at the structure of Z, we turn it into a numpy array to facilitate calculations
Z = Z.values

# Value added (last row of the first table)
VA = np.array([3.30, 22.4, 52.5])
# Final demand (last column)
Yfd = np.array([[1.9], [28.5], [47.8]])



X = np.concatenate((Z,Yfd), axis=1).sum(axis=1)

# Pasar a DataFrame
Z = pd.DataFrame(Z, columns = cs_names, index = cs_names )
Yfd = pd.DataFrame(Yfd, columns = cfd_names, index = cs_names )
Y = Yfd #solo hay un componente de la demanda final
VA = pd.DataFrame(VA, columns = ["VA"], index = cs_names)

X = pd.DataFrame(X, columns = ["O"], index = cs_names)




print("Z\n",Z,"\nVA\n",VA,"\nYfd\n",Y)
````

````python
# # Create object io
io = {"Z": Z, "Yfd": Yfd, "Y": Y, "VA": VA, "X": X}

# Metadata: dims
io["dims"] = {"C": C, "S": S, "FD": FD, "CS": CS, "CFD": CFD}

# Metadata: names
io["names"] = {"c_names": c_names, "s_names": s_names, "fd_names": fd_names,
               "cs_names": cs_names, "cfd_names": cfd_names}

print("Creado Objeto I-O")
````

````python
# ============================================================
# COMPROBACIÓN DE IDENTIDADES CONTABLES IO
# ============================================================
# Objetivo:
# Verificar que el output total por sector se obtiene de forma
# equivalente desde:
#   (1) el lado de los usos  : intermedios + demanda final
#   (2) el lado de los costes: intermedios + valor añadido
# ============================================================


# ------------------------------------------------------------
# 1. Output desde el lado de los USOS
#    X = Z·1  +  Y·1
# ------------------------------------------------------------

# Suma por filas de Z → ventas intermedias por sector
Z_sum_rows = Z.sum(axis=1)    # (cs,)

# Suma de la demanda final por sector
Y_sum = Y.sum(axis=1)         # (cs,)

# Output total por el lado de los usos
X_uses = (Z_sum_rows + Y_sum).to_numpy().squeeze()

print("=== Output por el lado de los USOS (Intermedios + Demanda Final) ===")
print(X_uses)


# ------------------------------------------------------------
# 2. Output desde el lado de los COSTES
#    X = 1·Z  +  VA
# ------------------------------------------------------------

# Suma por columnas de Z → compras intermedias por sector
Z_sum_cols = Z.sum(axis=0)    # (cs,)

# Output total por el lado de los costes
X_costs = (Z_sum_cols + VA.squeeze()).to_numpy().squeeze()

print("\n=== Output por el lado de los COSTES (Intermedios + Valor Añadido) ===")
print(X_costs)


# ------------------------------------------------------------
# 3. Comparación de ambos procedimientos
# ------------------------------------------------------------

diff = X_uses - X_costs

print("\n=== Chequeo de coherencia contable ===")
print(f"Suma de diferencias: {diff.sum():.3e}")
print(f"Máxima diferencia absoluta: {np.abs(diff).max():.3e}")

# Vector a vector (útil para debug)
print("\nOutput (Usos)  vs  Output (Costes):")
print(pd.DataFrame({
    "X_uses": X_uses,
    "X_costs": X_costs,
    "difference": diff
}))
````

````python
# ============================================================
# Leontief inverse (Demand-pull / Quantity model)
# ============================================================
# Objetivo:
# Construir la inversa de Leontief:
#
#     L = (I − A)^−1
#
# donde:
#   Z : matriz de flujos intermedios
#   X : vector de output total
#   A : matriz de coeficientes técnicos
#
# Interpretación:
# L captura los efectos directos + indirectos de un shock
# de demanda final sobre la producción.
# ============================================================


# ------------------------------------------------------------
# 0. Dimensiones y chequeos básicos
# ------------------------------------------------------------

Z = np.asarray(Z, dtype=float)
X = np.asarray(X, dtype=float).reshape(-1)

assert Z.shape[0] == Z.shape[1], "Z debe ser cuadrada"
assert Z.shape[0] == len(X),     "Dimensiones inconsistentes entre Z y X"
# ------------------------------------------------------------
# 1. Inversa diagonal del output (X̂⁻¹)
# ------------------------------------------------------------
# X̂ es la matriz diagonal del output.
# Usamos pseudo-inversa para evitar problemas si X contiene ceros.
#
# Económicamente:  divide cada columna j de Z por el output del sector j
# ------------------------------------------------------------

x_hat_inv = np.linalg.pinv(np.diag(X))


# ------------------------------------------------------------
# 2. Matriz de coeficientes técnicos: Cada columna j indica los inputs necesarios para producir una unidad de output del sector j
# ------------------------------------------------------------
# A_ij = Z_ij / X_j
# ------------------------------------------------------------

A = Z @ x_hat_inv


# ------------------------------------------------------------
# 3. Matriz identidad y sistema productivo
# ------------------------------------------------------------
# (I − A) representa los requerimientos netos del sistema
# ------------------------------------------------------------

I = np.eye(A.shape[0])
I_minus_A = I - A


# ------------------------------------------------------------
# 4. Inversa de Leontief
# ------------------------------------------------------------
# L = (I − A)^−1
#
# Si el sistema es productivo (ρ(A) < 1),
# esta inversa existe y es económicamente interpretable.
# ------------------------------------------------------------

L = np.linalg.inv(I_minus_A)

L
````

````python
# ============================================================
# Comprobación: x = L · y   (modelo quantity / demand-pull)
# ============================================================
# Objetivo: Verificar que el output total calculado a partir del modelo de Leontief coincide con el output observado X.
#
# Recordatorio teórico:
#   x = (I − A)^−1 · y  = L · y
#
# donde:
#   L : inversa de Leontief
#   y : demanda final TOTAL (agregada por sector)
# ============================================================


# ------------------------------------------------------------
# 1. Agregar la demanda final
# ------------------------------------------------------------
# Si Y es (sector × país demandante), primero hay que sumar por columnas
# ------------------------------------------------------------

y = Y.sum(axis=1)          # (cs,)

# ------------------------------------------------------------
# 2. Calcular el output inducido por el modelo
# ------------------------------------------------------------

x_calculated = (L @ y).squeeze()

# ------------------------------------------------------------
# 3. Comparación con el output observado
# ------------------------------------------------------------

diff = x_calculated - X

print("=== Chequeo x = L · y ===")
print(f"Suma de diferencias: {diff.sum():.3e}")
print(f"Máxima diferencia absoluta: {np.abs(diff).max():.3e}")

# Comparación sector a sector (útil para debug)
display(pd.DataFrame({
    "X_observed": X,
    "X_calculated": x_calculated,
    "difference": diff
}))

# “Small numerical differences arise due to floating-point arithmetic, matrix inversion, and minor imbalances in the input–output data, and are economically negligible.”
````

````python
# ============================================================
# MODELO DE PRECIOS DE LEONTIEF (COST-PUSH MODEL)
# ============================================================
# Objetivo: Calcular el vector de precios de equilibrio a partir de los costes primarios (valor añadido) y de la tecnología productiva del sistema input–output.
#
# Modelo teórico:
#   p = (I − A')⁻¹ · v
#
# donde:
#   A : matriz de coeficientes técnicos
#   p : vector de precios
#   v : coeficientes de valor añadido (costes primarios)
# ============================================================

# ------------------------------------------------------------
# 1. Transponer la matriz de coeficientes técnicos
# ------------------------------------------------------------
# En el modelo de precios, los costes se transmiten desde los inputs hacia los outputs, por lo que aparece A' en lugar de A.
# ------------------------------------------------------------

A_p = A.T

# ------------------------------------------------------------
# 2. Construir la inversa de Leontief en precios
# ------------------------------------------------------------
# L_p = (I − A')⁻¹
#
# Esta matriz recoge los efectos directos e indirectos de un aumento de costes primarios sobre los precios finales.
# ------------------------------------------------------------

I = np.eye(A_p.shape[0])
L_p = np.linalg.inv(I - A_p)


# ------------------------------------------------------------
# 3. Coeficientes de valor añadido
# ------------------------------------------------------------
# v_i = VA_i / X_i
#
# Representan el coste primario por unidad de producción (salarios, excedente, impuestos netos, etc.).
# Usamos np.divide para evitar problemas si X contiene ceros.
# ------------------------------------------------------------

# Agregar VA si tiene varias columnas
if isinstance(VA, pd.DataFrame):
    VA_vec = VA.sum(axis=1)
else:
    VA_vec = VA

# Convertir explícitamente a numpy
VA_vec = VA_vec.to_numpy()
X_vec  = np.asarray(X)

# Coeficientes de valor añadido (robusto a ceros)
v_coeff = np.divide(VA_vec, X_vec, out=np.zeros_like(X_vec, dtype=float), where=X_vec != 0)

# ------------------------------------------------------------
# 4. Vector de precios de equilibrio
# ------------------------------------------------------------
# p = L_p · v
#
# Precio que iguala costes intermedios + costes primarios en todo el sistema productivo.
# ------------------------------------------------------------

p = L_p @ v_coeff


# ------------------------------------------------------------
# 5. Normalización (opcional pero habitual)
# ------------------------------------------------------------
# El nivel de precios es indeterminado (numéraire).
# Normalizamos para que el precio medio sea 1.
# ------------------------------------------------------------

p_norm = p / p.mean()


# ------------------------------------------------------------
# 6. Resultados
# ------------------------------------------------------------

print("=== MODELO DE PRECIOS DE LEONTIEF ===")
print("\nCoeficientes de valor añadido (v):")
print(v_coeff)

print("\nVector de precios (sin normalizar):")
print(p)

print("\nVector de precios normalizado (media = 1):")
print(p_norm)


# ------------------------------------------------------------
# 7. Chequeo teórico fundamental
# ------------------------------------------------------------
# El sistema debe cumplir:
#   p = A' · p + v
# ------------------------------------------------------------

assert np.allclose(p, A.T @ p + v_coeff, atol=1e-6 ), "El sistema de precios no se cumple: p ≠ A'·p + v"
````

## The two-country production and trade system

The production system of two countries, \( s \) and \( r \), can be written as:

$$
\begin{bmatrix}
X_s \\
X_r
\end{bmatrix}
=
\begin{bmatrix}
A_{ss} & A_{sr} \\
A_{rs} & A_{rr}
\end{bmatrix}
\begin{bmatrix}
X_s \\
X_r
\end{bmatrix}
+
\begin{bmatrix}
Y_{ss} + Y_{sr} \\
Y_{rs} + Y_{rr}
\end{bmatrix}
$$

Rearranging the system yields:

$$
\begin{bmatrix}
X_s \\
X_r
\end{bmatrix}
=
\begin{bmatrix}
I - A_{ss} & - A_{sr} \\
- A_{rs} & I - A_{rr}
\end{bmatrix}^{-1}
\begin{bmatrix}
Y_{ss} + Y_{sr} \\
Y_{rs} + Y_{rr}
\end{bmatrix}
=
\begin{bmatrix}
B_{ss} & B_{sr} \\
B_{rs} & B_{rr}
\end{bmatrix}
\begin{bmatrix}
Y_s \\
Y_r
\end{bmatrix}
$$

---

### Notation

- \( B_{sr} \) denotes the \( N \times N \) block of the **Leontief inverse matrix**, capturing the total (direct and indirect) production requirements in country \( s \) associated with final demand in country \( r \).

- \( Y_s \) is an \( N \times 1 \) vector of **final products of country \( s \)**, defined as:
  
  $$
  Y_s = Y_{ss} + Y_{sr},
  $$
  
  where \( Y_{ss} \) represents **domestic final demand** in country \( s \), and   \( Y_{sr} \) denotes **exports of final products** from country \( s \) to country \( r \).

---

### Interpretation

This formulation highlights how production in each country depends not only on domestic final demand, but also on **cross-country input–output linkages**, which are fully captured by the block structure of the Leontief inverse.

-----------

$$
VB \;=\;
\left[
\begin{array}{cc}
V^{s} & V^{r}
\end{array}
\right]
\left[
\begin{array}{cc}
B^{ss} & B^{sr} \\
B^{rs} & B^{rr}
\end{array}
\right]
\;=\;
\left[
\begin{array}{cc}
V^{s} B^{ss} + V^{r} B^{rs} &
V^{s} B^{sr} + V^{r} B^{rr}
\end{array}
\right]
$$

=
\begin{bmatrix}
\hat{V}^{s} & 0 \\

\end{bmatrix}

B^{ss} & B^{sr} \\

\end{bmatrix}

\hat{Y}^{s} & 0 \\

\end{bmatrix}

\hat{V}^{r} B^{rs} \hat{Y}^{s} &
\hat{V}^{r} B^{rr} \hat{Y}^{r}
\end{bmatrix}
$$

### Decomposition of Value Added in Exports

The following expression decomposes the **domestic value added of country–sector \(s\)** embodied in its exports, following the logic of **Koopman–Wang–Wei / Borin–Mancini**.

### General expression

$$
\begin{aligned}
iE_s \;=\;
&\;\underbrace{V_s L_{ss} Y_{sr}}_{\text{(1) DVA}}
\;+\;
\underbrace{V_s L_{ss} A_{sr} L_{rr} Y_{rr}}_{\text{(2) VAX}} 
&\;+\;
\underbrace{
V_s L_{ss} A_{sr} L_{rr}
\sum_{j \neq r,s} Y_{rj}
}_{\text{(3) DC: exports to third countries}}
&\;+\;
\underbrace{
V_s L_{ss} A_{sr} L_{rr}
\sum_{j \neq r} A_{rj}
\sum_{k}
\sum_{l \neq s} B_{jk} K_{kl}
}_{\text{(4–6) DC: complex GVC circulation}}
\end{aligned}
\tag{A.16}
$$

---

#### General economic interpretation

The equation measures **how much value added generated in country \(s\)** is ultimately embodied in exports, whether it is:
- consumed directly abroad,
- transformed by other countries,
- or circulating through complex global value chains.

The whole expression starts from:

$$
V_s L_{ss}
$$

which represents the **domestic value added of country \(s\)** required to produce one unit of output in \(s\), taking into account **direct and indirect effects**.

---

### Decomposition by components

#### (1) DVA — Direct Domestic Value Added

- Country \(s\) produces final goods.
- These goods are exported **directly** to country \(r\).
- The value added from \(s\) is consumed in \(r\) without further transformation.

👉 This corresponds to the **domestic value added embodied in final goods exports**.

---

#### (2) VAX — Value Added Exported through Foreign Transformation

- Country \(s\) exports **intermediate inputs** to \(r\).
- Country \(r\) uses them to produce final goods.
- These goods are consumed in country \(r\) itself.

👉 The value added from \(s\) is exported **indirectly**, embodied in the production of \(r\).

---

#### (3–6) DC — Domestic Content in Global Value Chains

This block captures the **complex channels of global value chains (GVCs)**.

##### (3) Indirect exports to third countries

- \(s\) sells inputs to \(r\).
- \(r\) produces final goods.
- Those goods are exported to **third countries \(j\)**.

The value added from \(s\) is ultimately consumed outside \(r\).

##### (4–6) Complex circulation and domestic double counting

- The value added from \(s\) passes through multiple countries and sectors.
- It may cross several borders.
- It may partially return to the country of origin.
- It generates **domestic double counting** and **reflection effects**.

👉 This whole block is grouped under **Domestic Content (DC)**.

### Decomposición del valor añadido en las exportaciones

La siguiente expresión descompone el **valor añadido doméstico del país–sector \(s\)** incorporado en sus exportaciones, siguiendo la lógica de **Koopman–Wang–Wei / Borin–Mancini**.

### Expresión general

$$
\begin{aligned}
iE_s \;=\;
&\;\underbrace{V_s L_{ss} Y_{sr}}_{\text{(1) DVA}}
\;+\;
\underbrace{V_s L_{ss} A_{sr} L_{rr} Y_{rr}}_{\text{(2) VAX}} 
&\;+\;
\underbrace{
V_s L_{ss} A_{sr} L_{rr}
\sum_{j \neq r,s} Y_{rj}
}_{\text{(3) DC: exports to third countries}}
&\;+\;
\underbrace{
V_s L_{ss} A_{sr} L_{rr}
\sum_{j \neq r} A_{rj}
\sum_{k}
\sum_{l \neq s} B_{jk} K_{kl}
}_{\text{(4–6) DC: complex GVC circulation}}
\end{aligned}
\tag{A.16}
$$

---

#### Interpretación económica general

La ecuación mide **cuánto valor añadido generado en el país \(s\)** acaba incorporado en exportaciones, ya sea:
- consumido directamente en el extranjero,
- transformado por otros países,
- o circulando por cadenas globales de valor complejas.

Toda la expresión parte de:

$$
V_s L_{ss}
$$

que representa el **valor añadido doméstico del país \(s\)** necesario para producir una unidad de output en \(s\), teniendo en cuenta **efectos directos e indirectos**.

---

### Descomposición por componentes

#### (1) DVA — Domestic Value Added directo

- El país \(s\) produce bienes finales.
- Estos bienes se exportan **directamente** al país \(r\).
- El valor añadido de \(s\) se consume en \(r\) sin más transformaciones.

👉 Corresponde al **valor añadido doméstico contenido en exportaciones finales**.

---

#### (2) VAX — Value Added Exported vía transformación extranjera

- El país \(s\) exporta **inputs intermedios** a \(r\).
- El país \(r\) los utiliza para producir bienes finales.
- Dichos bienes se consumen en el propio país \(r\).

👉 El valor añadido de \(s\) se exporta **indirectamente**, incorporado en la producción de \(r\).

---

#### (3–6) DC — Domestic Content en cadenas globales de valor

Este bloque recoge los **canales complejos de las cadenas globales de valor (GVCs)**.

##### (3) Exportaciones indirectas a terceros países

- \(s\) vende inputs a \(r\).
- \(r\) produce bienes finales.
- Esos bienes se exportan a **terceros países \(j\)**.

El valor añadido de \(s\) acaba consumiéndose fuera de \(r\).

##### (4–6) Circulación compleja y doble conteo doméstico

- El valor añadido de \(s\) pasa por múltiples países y sectores.
- Puede cruzar varias fronteras.
- Puede reincorporarse parcialmente al país de origen.
- Genera **doble conteo doméstico** y **efectos de reflexión**.

👉 Todo este bloque se agrupa como **Domestic Content (DC)**.

---
