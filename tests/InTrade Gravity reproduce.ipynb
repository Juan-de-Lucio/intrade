```python
# Import lybraries
import intrade as it
import numpy as np
import pandas as pd
```

```python
# Create gravity object with gravity variables

grav_data = pd.read_csv("grav_data.csv")
grav_data = grav_data.dropna()
grav_data.exporter.unique()

dir_in = ""
obj = it.load_WIOD(dir_in,  2014, "WIOD 2016")

df = it.aggregate(obj)["df"]

obj_grav = it.obj_Grav(it.aggregate(obj))
obj_grav = obj_grav.merge(grav_data, on=["exporter", "importer"], how= "left")

# Drop ROW, no gravity data 
obj_grav = obj_grav[(obj_grav["exporter"] != "ROW") & (obj_grav["importer"] != "ROW")]
```

```python
# Denine parameters and estimations

# seleccion   = 1                                          # 3 different exercises: 1-increase, 2-asif, 3-inout
pais_ref    = "DEU"                                      # reference country
sigma       = 7                                          # sigma - elasticidad de sustitución
increase    = 15                                         # increase only for 1-increase

paises1 = ["AUT", "BEL","BGR","HRV","CYP","CZE","DNK","EST","FIN","FRA","DEU","GRC","IRL","HUN","ITA","LVA","LTU","LUX","MLT","POL","NLD","PRT","ROU","SVK","SVN","ESP","SWE"]
paises2 = ["USA"]
paises3 = ["GBR"]

# TRANSFORM VARIABLES
# LN de distancia
obj_grav['LN_DIST'] = np.log(obj_grav['DIST'])
variables = ["LN_DIST", "CNTG", "LANG","BRDR"]

# *********************
# ONLY GRAV ESTIMATES,
# *********************
(it.struc_grav(obj_grav,variables,pais_ref))


```

```python
# Contrafactua increase in trade cost

(it.struc_grav_incre(obj_grav, variables, paises1, paises2, pais_ref , increase, sigma))
```

```python
# A counterfactual in which one country is treated **as if** it were another

(it.struc_grav_asif(obj_grav, variables, paises1, paises2, paises3, pais_ref, sigma))
```

```python
# A counterfactual one country in/out a set of countries (eg. FTA)

(it.struc_grav_inout(obj_grav, variables, paises1, paises2, pais_ref, sigma))
```
